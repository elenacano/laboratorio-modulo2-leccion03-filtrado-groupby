```python
import pandas as pd
```

```python
# Carga el dataset de trabajo
df = pd.read_csv("datos/titulos_netflix_merged.csv")
pd.set_option("display.max_columns", None)
df.head()
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
condicion_anio = df["release_year"] == 2021
condicion_tipo = df["type"] == "Movie"
df_2021 = df[condicion_anio & condicion_tipo]
df_2021["title"].unique()
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
patron_regex = '^90'
filtro_duration = df["duration"].str.contains(patron_regex, na=False) 
df_90 = df[filtro_duration]
df_90[["title", "duration"]]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
patron_regex = 'India'
filtro_india= df["country"].str.contains(patron_regex, case=False, na=False) 
filtro_2018 = df["release_year"]>2018
df_india_2018 = df[filtro_india & filtro_2018]
df_india_2018[["title", "release_year", "country"]]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
condicion_calificacion = df["rating"].isin(['TV-14', 'R'])
df[condicion_calificacion][["title", "rating"]].reset_index()
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
patron_regex='Christopher Nolan|Martin Scorsese'
filtro_director= df["director"].str.contains(patron_regex, case=False, na=False)
df_director = df[filtro_director]
df_title_director = df_director[["title", "director"]]
df_title_director["director"].value_counts()
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
patron_regex="United States|United Kingdom|Canada"
filtro = df["country"].str.contains(patron_regex, case=False, na=False)
df_paises = df[filtro]

#df_paises[["title", "country"]]
df_paises
```

```python
varios_paises = df_paises["country"].value_counts()
varios_paises
```

```python
df_paises.shape
```

```python
patron_regex="^United States$|^United Kingdom$|^Canada$"
filtro_sing = df["country"].str.contains(patron_regex, case=False, na=False)
df_paises_sing = df[filtro_sing]
un_pais = df_paises_sing["country"].value_counts().sum()
un_pais
```

```python
varios_paises-un_pais
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
condicion_anios = df["release_year"].between(2015,2020)
df_anios_betwwen = df[condicion_anios]
df_anios_betwwen[["title", "release_year"]]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
df.head()
condicion_puntuacion = df["IMDB Score"].between(6.5,8.0)
filtro_nan = df["duration"].isna()
df_imdb = df[condicion_puntuacion & filtro_nan]
df_imdb[["title", "IMDB Score", "duration"]]
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración es de 90 o 150 minutos?

```python
patron = "^90|^150"
filtro_90_150= df["duration"].str.contains(patron, case=False, na=False)
df_90_150=df[filtro_90_150]
#pd.reset_option("display.max_rows")
#pd.set_option("display.max_rows", None)
df_90_150.head()
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
filtro = df["type"]=="TV Show"
df_tv_show = df[filtro]

patron = 'Love '
condicion_love = df["title"].str.contains(patron, case=False, na=False)
df_tv_show_love = df_tv_show[condicion_love]
df_tv_show_love.head(10)
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
patron="\w\sJohnson"
filtro_johnson = df["director"].str.contains(patron, case=False, na=False)
df_johnson = df[filtro_johnson]
df_johnson.head()

```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
patron="Documentaries"
filtro_documentaries = df["listed_in"].str.contains(patron, case=False, na=False)

patron2="Mike Tyson"
filtro_tyson = df["cast"].str.contains(patron2, case=False, na=False)


df_johnson = df[filtro_documentaries & filtro_tyson]
df_johnson.head()
```

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python
condicion_anios = df["release_year"]<2010
condicion_tipo = df["type"] == "Movie"
df_2010 = df[condicion_anios & condicion_tipo]
df_2010[["title", "release_year", "type"]]
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
df_imdb = df.nlargest(10, 'IMDB Score').filter(["title", "IMDB Score"])
df_imdb
```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

```python
df_imdb = df.nsmallest(10, 'IMDB Score').filter(["title", "IMDB Score"])
df_imdb
```

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.
